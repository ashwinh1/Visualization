### Dashboard Application

```python
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1('My First Dash Project',\
            style={'text-align': 'center','font-family':'Arial'}),\
    
    dcc.Dropdown(id='genre-choice',
                 options=[{'label':x,'value':x} for x in sorted(df.Genre.unique())],
                value='Action',style={'width':'100%','font-family':'Arial'}),\
    
    dcc.Graph(id='my-graph',figure={}),
    ])

@app.callback(
    Output(component_id='my-graph',component_property='figure'),
    Input(component_id='genre-choice',component_property='value')
    )

def interactive_graphing(value_genre):
#     print(value_genre)
    dff = df[df.Genre==value_genre]
    fig = px.histogram(data_frame=dff,
             x='World Sales',
             y='Platform',
             color_discrete_sequence=['blue'],
             width=800,
             labels=True,
             height=500).update_layout(font_family="Arial")\
            .update_layout(yaxis={'categoryorder':'total ascending'},xaxis_title='Total Sales')
    return fig

if __name__=='__main__':
    app.run_server()
```
